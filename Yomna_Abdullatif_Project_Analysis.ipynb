```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales = pd.read_csv('/content/drive/MyDrive/sales_data_sample.csv', encoding='latin-1')
sales.info()
```

```python
sales.isnull().sum()
```

```python
sales = sales.dropna(how = 'all')
```

```python
sales['ORDERDATE'] = pd.to_datetime(sales['ORDERDATE'])
```

```python
sales.info()
sales.head(20)
```

```python
sales.describe()
```

```python
sales['SALES'].describe()
```

```python
Customer_segment = []
s = sales['SALES']
for sale in s :
  if sale <= 2203:
    Customer_segment.append('Low')
  elif 2203 < sale <= 3184:
    Customer_segment.append('Medium')
  else:
    Customer_segment.append('High')
sales['Customer_segment'] = Customer_segment

df_customer_segment = pd.DataFrame({'Customer_segment': Customer_segment})
df = pd.merge(sales, df_customer_segment, left_index=True, right_index=True)
```

```python
df.drop(['Customer_segment_y'], axis=1, inplace = True)
df.rename({'Customer_segment_x': 'Customer segment'}, axis=1, inplace = True)
df.head(1)
```

```python
new = df['PRODUCTCODE'].str.split('_', expand=True)
df_1 = pd.concat([df, new], axis=1)
df_1.drop(0, axis=1, inplace = True)
df_1.drop('ADDRESSLINE2', axis=1, inplace = True)
df_1.drop(['PRODUCTCODE'], axis=1, inplace = True)
df_1.rename({1: 'PRODUCTCODE'}, axis=1, inplace = True)
df_1.head(1)
```
